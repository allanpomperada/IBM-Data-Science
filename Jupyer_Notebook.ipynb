### I would be doing some simple BMI calculation with two sets of list.


```python
height = [1.63, 1.59, 1.60]
weight = [56, 63, 59]


import numpy as np

np_height = np.array(height)
np_weight = np.array(weight)

# BMI = weight in kg / (height in m ^2)

bmi = np_weight / (np_height ** 2)


print(bmi)
```

From the given data

height|weight
---|---
1.63|56
1.59|63
1.60|59

> The result gives us the following BMI respectively:
- 21.07719523
- 24.91990032
- 223.046875
